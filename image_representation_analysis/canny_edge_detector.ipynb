# Canny Edge Detector

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/sunflower.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

### Implement Canny Edge Detection

* We used the OpenCV function Canny, which is well-documented, [here](http://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_canny/py_canny.html).

```python
# Define lower and upper thresholds for hysteresis

# It is recommended to use a low to high threshold ratio between 1:2 or 1:3 for best results.
lower = 120
upper = 240

# Canny detects the edges of the flower and produces a binary image
# Canny function returns an edge detected image
edges = cv2.Canny(gray, lower, upper)

plt.imshow(edges, cmap='gray')
```

```python
# Try Canny using "wide" and "tight" thresholds

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 180, 240)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
```

The wide image shows a lot of detected edges, and the tight shows fewer since the cutoff for week edges is hight. In both cases, we can see that Canny detects boundaries quite well.
